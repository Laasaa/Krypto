```python
# Import library for cryptocompare data
import cryptocompare

# Import data manipulation library
import pandas as pd
import pandas_ta as pta
# Import datetime package
from datetime import datetime

# Import matplotlib and set the style for plotting
import matplotlib.pyplot as plt

# Fetch the raw ticker list
#raw_ticker_data = cryptocompare.get_coin_list()

# Convert the raw data from dictionary format to DataFrame
#all_tickers = pd.DataFrame.from_dict(raw_ticker_data).T

while True:

    valinta = int(input(" Lista kryptovaluuttoja(1)\n Vaaluutan nykyinen arvo euroissa(2)\n Vertailla valuuttojen arvoja(3)"))

    if valinta == 1:
      df = pd.read_csv('coins.csv')
      print(df)

    if valinta == 2:
      val_nimi = str(input("Anna haluamisi valuutan lyhenne (Esim. ETH,BTC,USD): "))
      val_hinta = cryptocompare.get_price(val_nimi)
      print("Valuutan arvo euroissa:",val_hinta)      
          
    if valinta == 3:
        kol = str (input("Anna ensimmäisen valuutan lyhenne (Esim. ETH,BTC,USD): "))
        kol2 = str (input("Anna toisen valuutan lyhenne (Esim. ETH,BTC,USD): "))
        valinta_aika = str (input("Datapointtien etäisyys (minute, hour, day): "))
        if valinta_aika == "minute":

          # Fetch the raw price dataraw_price_data = \
            raw_price_data = \
            cryptocompare.get_historical_price_minute(kol, kol2, limit=2000, exchange='CCCAGG', toTs=datetime(2023, 1, 1, 0, 0))
            print(raw_price_data)
          # Convert the raw price data into a DataFrame
            minute_price_data = pd.DataFrame.from_dict(raw_price_data)

          # Set the time columns as index and convert it to datetime
            minute_price_data.set_index("time", inplace=True)
            minute_price_data.index = pd.to_datetime(minute_price_data.index, unit='s')
       
      
          # Plot the close price
            plt.figure(figsize=(15, 7))
            plt.style.use('ggplot')
            plt.plot(minute_price_data.close)

          # Set title and labels for the plot
            plt.title("Valuutan kurssi", fontsize=16)
            plt.style.use('ggplot')
            plt.xlabel('Kuukausi,päivä,tunti', fontsize=15)
            plt.ylabel('Valuuttojen suhde', fontsize=15)
            plt.tick_params(axis='both', labelsize=15)

          # Show the plot
            plt.show()
            break
            
        elif valinta_aika == "hour":
          # Fetch the raw price dataraw_price_data = \
            raw_price_data = \
            cryptocompare.get_historical_price_hour(kol, kol2, limit=2000, exchange='CCCAGG', toTs=datetime(2023, 1, 1, 0, 0))
           
          # Convert the raw price data into a DataFrame
            hourly_price_data = pd.DataFrame.from_dict(raw_price_data)

          # Set the time columns as index and convert it to datetime
            hourly_price_data.set_index("time", inplace=True)
            hourly_price_data.index = pd.to_datetime(hourly_price_data.index, unit='s')

          # Plot the close price
            plt.figure(figsize=(15, 7))
            plt.style.use('ggplot')
            plt.plot(hourly_price_data.close)

          # Set title and labels for the plot
            plt.title("Valuutan kurssi", fontsize=16)
            plt.style.use('ggplot')
            plt.xlabel('Vuosi,kuukausi,päivä', fontsize=15)
            plt.ylabel('Valuuttojen suhde', fontsize=15)
            plt.tick_params(axis='both', labelsize=15)
            
          # Show the plot
            plt.show()
            break
            
        elif valinta_aika == "day":
          # Fetch the raw price dataraw_price_data = \
            raw_price_data = \
            cryptocompare.get_historical_price_day(kol, kol2, limit=2000, exchange='CCCAGG', toTs=datetime(2023, 1, 1, 0, 0))

          # Convert the raw price data into a DataFrame
            day_price_data = pd.DataFrame.from_dict(raw_price_data)
            
          # Set the time columns as index and convert it to datetime
            day_price_data.set_index("time", inplace=True)
            day_price_data.index = pd.to_datetime(day_price_data.index, unit='s')
       
          # Plot the close price
            plt.figure(figsize=(20, 10))
            plt.plot(day_price_data.close)
            
            
          # Set title and labels for the plot
            plt.title("Valuutan kurssi", fontsize=16)
            plt.style.use('ggplot')
            plt.xlabel('Vuosi', fontsize=15)
            plt.ylabel('Valuuttojen suhde', fontsize=15)
            plt.tick_params(axis='both', labelsize=15)
    
          # Show the plot
            plt.show()
            break
         
            
```
